### ***KPI's Requiredment***

```python
# Total Sales
total_sales = df['Sales'].sum()

# Average Sales
avg_sales = df['Sales'].mean()

# Number of items sold
no_of_items_sold = df['Sales'].count()

# Average rating                               
avg_rating = df['Rating'].mean()

# Display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.2f}")
print(f"Number of Items Sold: {no_of_items_sold:,}")
print(f"Average Rating: {avg_rating:.2f}")
```

### ***chart reqirement***

### *total sales by fat content*

```python
import matplotlib.pyplot as plt

# Group sales by fat content
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()  # note the ()

# Plot pie chart
plt.pie(
    sales_by_fat, 
    labels=sales_by_fat.index,  # comma here
    autopct='%.2f%%',
    startangle=90
)
```

### *total sales by item type*

```python
 #Total sales by Item Type
sales_by_item_type = df.groupby("Item Type")["Sales"].sum().sort_values(ascending=False)

# Display table
print("Total Sales by Item Type:/n")
print(sales_by_item_type)

# Visualization (Bar Chart)
plt.figure(figsize=(10,6))
sales_by_item_type.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title("Total Sales by Item Type", fontsize=14, weight='bold')
plt.xlabel("Item Type")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### *fat content by outlet for total sales*

```python
sales_by_fat_outlet = df.groupby(["Item Fat Content", "Outlet Identifier"])["Sales"].sum().reset_index()

# Pivot for bar chart
pivot_data = sales_by_fat_outlet.pivot(index="Outlet Identifier", columns="Item Fat Content", values="Sales")

# Plot bar chart
pivot_data.plot(kind="bar", figsize=(12,6))

plt.title("Total Sales by Fat Content across Outlets", fontsize=14, weight="bold")
plt.xlabel("Outlet Identifier")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Fat Content")
plt.tight_layout()
plt.show()
```

### *total sales by outlet establishment*

```python
import matplotlib.pyplot as plt

# Group by Outlet Establishment Year and sum sales
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# Create figure
plt.figure(figsize=(9,5))

# Plot line chart with markers
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment')

# Add data labels on top of each point
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

### *sales by outlet size*

```python
# Step 1: Group total sales by Outlet Size
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

# Step 2: Create pie chart
plt.figure(figsize=(6,6))
explode = [0.05] * len(sales_by_size)  # slightly separate each slice

plt.pie(
    sales_by_size, 
    labels=sales_by_size.index,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    explode=explode
)

plt.title('Sales Distribution by Outlet Size', fontsize=16)
plt.axis('equal')  # ensures pie is circular
plt.show()
```

### *sales by outlet location type*

```python
# Step 1: Group total sales by Outlet Location Type
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum()

# Step 2: Plot bar chart
plt.figure(figsize=(6,5))
plt.bar(sales_by_location.index, sales_by_location.values, color='lightgreen', edgecolor='black')

# Add titles and labels
plt.title('Total Sales by Outlet Location Type', fontsize=16)
plt.xlabel('Outlet Location Type', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=0)

# Optional: show sales numbers on top of each bar
for i, value in enumerate(sales_by_location.values):
    plt.text(i, value + 50, str(round(value,2)), ha='center', fontsize=10)

plt.show()
```
